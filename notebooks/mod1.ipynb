## A Program That Saves Values

Recall that making a function call replaces that call with whatever the function returns. These lines do not do anything with that return, so we don't see any value if they are just on a line. 

```
double(5)
double(17)
double(-3)
```

Instead, save the results of the function call to a variable. Then you can print it, or use it future calculations.



````python
def double(x):
  """ x is a number
  Returns the double of x """
  return x*2

double(5)
double(17)
double(-3)

n1 = double(5)
n2 = double(17)
n3 = double(-3)
n4 = double(n1)

print(n1, n2, n3, n4)
````

## Working With Types and Input

Casting is the process of making a new version of an object that is of a different type. IN the cell below, two steps are being done in one in the line

```
 int(input("Your age: "))
 ```
 
 The first is to get the input from the user, and save it as a string. 
 
 ```
 input("Your age: ")
 ```
 
 Then, that string value is converted to the integer version of whatever was typed in. 
 
 * If the user typed in `9`: `int("9")`  <-- Python knows how to do this
 * If the user typed in `hi`: `int("hi")`  <-- Python gives an error with this

````python
num1 = input("Your age: ")
print(5*num1)

num2 = int(input("Your age: "))
print(5*num2)
````
